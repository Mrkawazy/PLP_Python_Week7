# Importing the libraries 

```python
#Importing the pandas library to work with dataframes
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Task 1: Load and Explore the Dataset

```python
try:
    #Reading the data from the CSV file into a pandas dataframe
    df=pd.read_csv("supermarket_sales - Sheet1.csv")
    #Getting snap of the data set for 5 records
    print('Dataset Loaded Successfully')
except ImportError:
    print("Data file not found. Please ensure the CSV file is in the correct directory.")
except pd.errors.ParserError:
    print("Error parsing the CSV file. Please check the file format.")
```

```python
#Checking the data types of the columns in the dataframe
df.info()
#Checking for missing values in the dataframe
df.isna().sum()
```

# Task 2: Basic Data Analysis

```python
# Compute the basic statistics of the numerical columns
df.describe()
```

```python
df.groupby('City').agg({'Total':'mean','Unit price':'mean','Quantity':'mean'})
#Grouping the data by 'City' and calculating the mean of the 'Total' column
```

```python
# Set the figure size for the plot
plt.figure(figsize=(10, 5))

# Create a bar plot to visualize total sales by product line
# Use the sum of 'Total' for each 'Product line' and apply a blue color palette
sns.barplot(x='Product line', y='Total', data=df, estimator=sum, ci=None, palette='Blues_d')

# Set the x-axis label
plt.xlabel('Product Line')

# Set the y-axis label
plt.ylabel('Total Sales')

# Set the title of the plot
plt.title('Total Sales by Product Line')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Create a pie chart to visualize the distribution of branches
df['Branch'].value_counts().plot(
    kind='pie',  # Specify the chart type as pie
    figsize=(10, 5),  # Set the figure size
    autopct='%1.1f%%',  # Display percentages on the pie chart
    startangle=90,  # Start the pie chart at 90 degrees
    wedgeprops={'width': 0.3}  # Adjust the width of the pie slices
)

# Remove the default y-axis label
plt.ylabel('')

# Set the title of the pie chart
plt.title('Gender Distribution')

# Add a legend to the chart, showing the unique branch labels
plt.legend(loc='upper right', labels=df['Branch'].unique())

# Ensure the pie chart is circular by setting an equal aspect ratio
plt.axis('equal')

# Display the pie chart
plt.show()
```

# Task 3: Data Visualization

```python
# Plotting total sales over time as a line plot
df.groupby('Date')['Total'].sum().plot(kind='line', figsize=(10, 5), color='green')
plt.xlabel('Date')  # Label for the x-axis
plt.ylabel('Total Sales')  # Label for the y-axis
plt.title('Total Sales Over Time')  # Title of the plot
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.show()

# Plotting gross income by gender as a bar chart
df.groupby('Gender')['gross income'].sum().plot(kind='bar', figsize=(10, 5), color='purple')
plt.xlabel('Gender')  # Label for the x-axis
plt.ylabel('Gross Income')  # Label for the y-axis
plt.title('Gross Income by Gender')  # Title of the plot
plt.show()

# Creating a scatter plot to visualize the relationship between 'Unit price' and 'Total'
plt.figure(figsize=(10, 5))  # Set the figure size
plt.scatter(df['Unit price'], df['Total'], alpha=0.5, color='orange')  # Scatter plot with transparency
plt.title('Scatter Plot of Unit Price vs Total')  # Title of the plot
plt.xlabel('Unit Price')  # Label for the x-axis
plt.ylabel('Total')  # Label for the y-axis
plt.grid(True)  # Add grid lines to the plot
plt.show()

# Creating a histogram to visualize the distribution of 'Unit price'
plt.figure(figsize=(10, 5))  # Set the figure size
plt.hist(df['Unit price'], bins=20, color='skyblue', edgecolor='black')  # Histogram with 20 bins
plt.title('Histogram of Unit Price')  # Title of the histogram
plt.xlabel('Unit Price')  # Label for the x-axis
plt.ylabel('Frequency')  # Label for the y-axis
plt.grid(axis='y', alpha=0.75)  # Add grid lines to the y-axis
plt.show()
```

```python
# Calculate the total revenue
total_income = df['gross income'].sum()

# Find the best-selling product (based on Quantity Sold)
best_selling_product = df.groupby('Product line')['Quantity'].sum().idxmax()
best_selling_quantity = df.groupby('Product line')['Quantity'].sum().max()

# Identify the day with the highest sales (total revenue)
df['Date'] = pd.to_datetime(df['Date'])
daily_sales = df.groupby('Date')['Total'].sum()
highest_sales_day = daily_sales.idxmax()
highest_sales_value = daily_sales.max()

# Save the results to a new file called sales_summary.txt
with open('sales_summary.txt', 'w') as f:
    f.write(f'Total Income: ${total_income}\n')
    f.write(f'Best-Selling Product: {best_selling_product} ({best_selling_quantity} units sold)\n')
    f.write(f'Highest Sales Day: {highest_sales_day.date()} (${highest_sales_value})\n')

# Print the insights in a user-friendly format
print(f'Total Income: ${total_income}')
print(f'Best-Selling Product: {best_selling_product} ({best_selling_quantity} units sold)')
print(f'Highest Sales Day: {highest_sales_day.date()} (${highest_sales_value})')
```
